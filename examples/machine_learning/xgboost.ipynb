Predictor = features; response = target. 

Conditions to use Poisson regression:

1) Response is count data (non negative integers: 0, 1, 2, 3, 4, etc)

2) Each observation is independent of another one. One observation should not provide information of another one.

3) The distribution of counts follows a Poisson distributions, meaning the expected and the count are similar. One way to test is to plot the expected and the observed and see they are similar. (Hmm, but how to know the expected, as this is what we try to predict?!)

4) The mean and variance of the model are equal. Property of the Poisson distribution. Called equidisperion. Often this is violated and have overdispersion.

Typical of Poisson distributions, most have zero.
